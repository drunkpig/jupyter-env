```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("csv.csv",encoding='gbk')
```

```python
df
```

```python

df.columns
```

```python
df2 = df[['代码', '名称',  '本期.5', '上年同期(调整后).5',
       '增长率(%).5',  '证监会(新)', '同花顺(新)', '申万行业', '定期报告实际披露日期']]
df2
```

```python

# 净利润是一季度的，要乘以4

# 不要的：银行、，游戏，旅游，酒店，机场，房地产，
# 利润<1亿、ST股,04-29才公布业绩的，利润负增长的不要

# 要的：市盈率<40
df3 = df2.rename(columns={"代码":"code","名称":"name", "上年同期(调整后).5":"净利","增长率(%).5":"增长率", "同花顺(新)":"同花顺","证监会(新)":"证监会"})
df3
```

```python


exclude_key = ['银行','游戏','旅游','酒店','机场','房地产',]

for k in exclude_key:
    df3 = df3[~df3['证监会'].str.contains(k)]
    df3 = df3[~df3['同花顺'].str.contains(k)]
    df3 = df3[~df3['申万行业'].str.contains(k)]
    
df3
```

```python
df4 = df3[df3.定期报告实际披露日期!='2022-04-29']
df4
```

```python
df5 = df4[~df4['name'].str.contains('ST')]
df5
```

```python
# 增长率<0 # df.apply(lambda x: x.str.replace(',','')).astype(np.int)
#df5[['增长率',"净利"]] = df5[["增长率","净利"]].apply(lambda x: x.str.replace(',','')).astype(np.float)
df6 = df5[df5.增长率>0]
df6
```

```python
# 利润<1亿
df7 = df6[df6.净利>10000]
df7
```

```python
df7.to_csv("./result.csv",encoding='utf8')
```

```python
import qgrid
w = qgrid.show_grid(df7)
w
```

```python
df_shizhi = pd.read_csv("shizhi.csv",encoding='gbk')
df_shizhi
```

```python
df_temp = df7.copy()

df_temp['股票代码'] = df_temp['code']
df_temp
```

```python
result = pd.merge(df_temp, df_shizhi, how='inner',  on='股票代码')

result
```

```python
result['市盈率'] = result["总市值(元)2022.04.29"]/result['净利']/10000
result
```

```python
ok = result[result.市盈率<40]
ok.to_csv("result.csv", encoding='utf-8')
```
